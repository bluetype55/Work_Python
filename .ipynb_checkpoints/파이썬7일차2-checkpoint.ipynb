# subflot 여러개의 플롯을 한번에 보여줌

```python
fig = plt.figure(figsize = (16,10),dpi = 80)
grid = plt.GridSpec(4,4, hspace = 0.5, wspace = 0.2)

ax_main = fig.add_subplot(grid[:-1, :-1])
ax_right = fig.add_subplot(grid[:-1,-1], xticklabels = [], yticklabels = [])
ax_bottom = fig.add_subplot(grid[-1,0:-1],xticklabels=[],yticklabels=[])

ax_main.scatter('NO2','PM10',color = 'r',data= df)
ax_right.hist(df.PM10,40,orientation='horizontal',color='lightblue')
ax_bottom.hist(df.NO2,40,orientation='vertical',color='lightpink')
```

```python
tmp_df
```

```python
my_df = tmp_df[tmp_df.teamPosition == 'JUNGLE']
```

```python
my_df
```

```python
xdata = my_df.kills
ydata = my_df.totalDamageDealtToChampions
```

```python
plt.figure()
plt.plot(xdata,ydata,color='y',marker='o',linestyle='None')
plt.xlabel('킬')
plt.ylabel('딜량')
plt.title('딜량에 따른 킬 횟수와의 관계')
plt.grid(True)
```

```python
ydata2 = my_df.totalDamageTaken
```

```python
plt.figure(figsize=(15,10))
plt.plot(xdata,ydata,color='g',marker='^',linestyle='None')
plt.plot(xdata,ydata2,color='r',marker='v',linestyle='None')
plt.xlabel('킬')
plt.ylabel('딜량, 피해량')
plt.title('킬수에 따른 딜량과 피해량')
plt.grid(True)
```
