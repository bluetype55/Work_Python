### 숫자, 계산할 숫자, 계산 부호를 넣어서 함수를 만들기

```python
def calc(n,sign,b):
    if sign == '+':
        res = n+b
        return res
    elif sign == '-':
        res=n-b
        return res
    elif sign == '*':
        res=n*b
        return res
    elif sign == '/':
        res=n/b
        return res
    else:
        print('잘못된 입력입니다')
        return False
```

```python
def comp():
    while True:
        sum1=[]
        sub=[]
        mul=[]
        div=[]
        math=input('수식을 입력해주세요')
        if '+' in math:
            sum1=math.split('+')
            print(math+'=',int(sum1[0])+int(sum1[1]))
        elif '-' in math:
            sub=math.split('-')
            print(math+'=',int(sub[0])-int(sub[1]))
        elif '*' in math:
            sub=math.split('*')
            print(math+'=',int(sub[0])*int(sub[1]))
        elif '/' in math:
            sub=math.split('/')
            print(math+'=',int(sub[0])/int(sub[1]))
        else: print('수식을 잘못입력하셨습니다')
```

```python
comp()
```

# DamageDealt = 준 피해량
# DamageTaken = 받은 피해량
# DamageDealt/DamageTaken ->컬럼 만들기
# kda만들기

```python
new_lol['deal_ratio']='None'
new_lol['kda']='None'
new_lol
```

```python
def kda(x):
    try:
        kda = round((x.kills+x.assists)/x.deaths,2)
        return kda
    except:
        kda= x.kills+x.assists
        return kda
```

```python
def dd(x):
    try:
        d1 = round((x.damageDealt/x.damageTaken),2)
        return d1
    except:
        d1= x.damageDealt
        return d1
```

```python
new_lol['deal_ratio']=new_lol.damageDealt/new_lol.damageTaken
```

```python
new_lol['deal_ratio']=new_lol.apply(lambda x: dd(x),axis=1)
```

```python
new_lol['kda'] = new_lol.apply(lambda x: kda(x), axis=1)
```

```python
new_lol.sort_values(by=['deal_ratio'],ascending=False)
```
